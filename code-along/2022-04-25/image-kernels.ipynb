```python
image = plt.imread(f'{data_dir}/{os.listdir(data_dir)[0]}')
image.shape
```

```python
resize_factor = .5
new_shape = (int(image.shape[1]*resize_factor), int(image.shape[0]*resize_factor)) # width, height

image_resized = cv2.resize(image, new_shape)
print(image_resized.shape)
plt.imshow(image_resized);
```

## Kernels (Filters)

```python
ridge_filter = np.array([
                        [0, -1, 0], 
                        [-1, 4, -1], 
                        [0, -1, 0]])

blur_filter = np.array([
                        [.0625, .125, .0625], 
                        [.125, .25, .125], 
                        [.0625, .125, .0625]])

outline_filter = np.array([
                        [-1, -1, -1], 
                        [-1, 20, -1], 
                        [-1, -1, -1]])

filters = [ridge_filter, blur_filter, outline_filter]
filter_names = ['ridge', 'blur', 'outline']
```

```python
processed_image_path = f'{data_dir}/processed_images'
# try:
#     os.mkdir(processed_image_path)
# except: FileExistsError as file_error:
#     print(file_error)
```

```python
processed_image_path = f'{data_dir}/processed_images'
try:
    os.mkdir(processed_image_path)
except FileExistsError as file_error:
    print(file_error)
```

```python
fig, axes = plt.subplots(1,len(filters))

for ax, filter, filter_name in zip(axes, filters, filter_names):
    filtered_image = cv2.filter2D(image_resized, -1, kernel=filter)
    ax.imshow(filtered_image)
    ax.axis('off')
    ax.set(title=f'{filter_name} kernel')
```

```python
fig.savefig(f'{processed_image_path}/filtered_images.png')
```
